# Quantum Teleportation
  Quantum Teleportation is a method for transfering state of quantum particle. The information is not actually copied, it's just extracted. Information of the particle that we copied is end up changes. Firstly, we will see how does Quantum Teleportation occures, then try to violate **No-Clone Theorem** .
 ### Necessary Imports

```python
from qiskit import QuantumCircuit, QuantumRegister, execute,Aer
from qiskit.visualization import plot_bloch_multivector
```

### Creating Backend

```python
be=Aer.get_backend("statevector_simulator")
```

### Teleporting State
    Let's teleport the following state

```python
qc=QuantumCircuit(1)
qc.initialize(((2/3)**.5,-((1/3)**.5)*1.j),0)
sv=execute(qc,be).result().get_statevector()
display(plot_bloch_multivector(sv))
```

```python
qc=QuantumCircuit(4)
qc.initialize(((2/3)**.5,-((1/3)**.5)*1.j),0)
sv=execute(qc,be).result().get_statevector()
display(plot_bloch_multivector(sv))

qc.h(1)
qc.cx(1,2)


qc.cx(0,1)
qc.h(0)
qc.cz(0,2)

qc.cx(1,2)

sv=execute(qc,be).result().get_statevector()
display(plot_bloch_multivector(sv))
display(qc.draw("mpl"))
```

As you can see state of qubit0 teleported to state of qubit2 and information of qubit0 has been lost.

### Let's try to violate No-Clone Theorem

```python
qc=QuantumCircuit(4)
qc.initialize(((2/3)**.5,-((1/3)**.5)*1.j),0)
sv=execute(qc,be).result().get_statevector()
display(plot_bloch_multivector(sv))

qc.h(1)
qc.cx(1,2)
qc.cx(1,3)


qc.cx(0,1)
qc.h(0)
qc.cz(0,2)
qc.cz(0,3)

qc.cx(1,2)
qc.cx(1,3)

sv=execute(qc,be).result().get_statevector()
display(plot_bloch_multivector(sv))
display(qc.draw("mpl"))
```

We tried to teleport q0 to both q2 and q3, but we didn't accomplish that. We lost information of both q0, q1 and q3
